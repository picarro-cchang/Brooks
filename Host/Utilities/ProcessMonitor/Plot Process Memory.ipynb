```python
import cPickle
fname = 'ProcessMonitor_1382223770.pic'
with file(fname,'rb') as fp:
    while True:
        try:
            now, procByPort = cPickle.load(fp)
            line = ['%10d' % now]
            for p in sorted(procByPort.keys()):
                proc = procByPort[p]
                line.append('%6dK' % (proc['rss']//1024))
            print ''.join(line)
        except Exception as e:
            break
```
