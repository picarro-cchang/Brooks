## Converting position to latitude and longitude

If $R_e$ is the radius of the earth and $(N,E)$ are the coordinates of a point relative to some reference latitude ($\varphi$) and longitude ($\lambda$), the latitude and longitude of the point are
$$\varphi' = \varphi + N/R_e$$
$$\lambda' = \lambda + E/(R_e\cos\varphi)$$

We represent the point $(N,E)$ by the complex number $z=N+\mathrm{j}E$.

## Simulating Path of Car and Wind

```python
import MobileKit.AnalyzerServer.SwathProcessor as sp
from numpy import *
from collections import namedtuple

Re = 6378100
RTD = 180.0/pi
DTR = pi/180.0

def ne2latlng(lat0,lng0,N,E):
    # Calculte the latitude and longitude at point(s) a distance N north and  E east from the reference position lat0, lng0
    #  All latitudes and longitudes are in degrees
    lat = lat0 + RTD * (N / Re)
    lng = lng0 + RTD * (E / (Re * cos(DTR * lat)))
    return namedtuple("LatLng", ["lat", "lng"])(lat, lng)
    
# Data required by the swath processor
DataTuple = namedtuple("DataTuple", ["EPOCH_TIME", "GPS_ABS_LAT", "GPS_ABS_LONG", "GPS_FIT", 
                                     "WIND_N", "WIND_E", "WIND_DIR_SDEV", "CAR_SPEED", 
                                     "ValveMask", "INST_STATUS"])

def makeData(r, dt, tstart, speed, wspeed, wdir, wsdev, lat0, lng0):
    # r is the radius
    # dt is the intersample time
    # tstart is the start epoch time
    t = dt * arange(ceil(4*pi*r/speed))
    theta = speed * t / r
    z = r * exp(1j*theta)
    # wind is from a single direction
    w = wspeed * exp(1j*DTR*wdir)
    #
    lat, lng = ne2latlng(lat0, lng0, real(z), imag(z))
    # Indicate that all GPS data are good
    fit = 2 * ones(len(t))
    windN = real(w) * ones(len(t))
    windE = imag(w) * ones(len(t))
    windSdev = wsdev * ones(len(t))
    carSpeed = speed * ones(len(t))
    valveMask = zeros(len(t))
    instStatus = 963 * ones(len(t))
    return DataTuple(tstart+t,lat,lng,fit,windN,windE,
                     windSdev,carSpeed,valveMask,instStatus)
r = 100.0
dt = 1.0
tstart = 1000000000
speed = 10.0
wspeed = 5.0
wdir = 45.0
wsdev = 30.0
lat0 = 37.4 
lng0 = -122.0

d = makeData(r, dt, tstart, speed, wspeed, wdir, wsdev, lat0, lng0)
plot(d.GPS_ABS_LONG, d.GPS_ABS_LAT, '.')
axis('equal')
grid(True)
```

The data need to be rearranged as a list of dictionaries before being passed to the swath processor. This can be conveniently done using a generator

```python
def asSource(dataTuples):
    f = dataTuples._fields
    for x in zip(*dataTuples):
        yield dict(zip(f,x))
```

```python
maxWindow = 10
stabClass = "D"
minLeak = 1.0
minAmpl = 0.05
astdParams = {"a": 0.15*pi,"b": 0.25,"c": 0.0}
source = [s for s in asSource(d)]
r = sp.process(source,maxWindow,stabClass,minLeak,minAmpl,astdParams)
for k in r:
    r[k] = asarray(r[k])
print r
```

```python
plot(r["GPS_ABS_LONG"], r["GPS_ABS_LAT"], '.', r["GPS_ABS_LONG"]+r["DELTA_LONG"], r["GPS_ABS_LAT"]+r["DELTA_LAT"], '.')
axis('equal')
grid(True)
```
