```python
from numpy import *
from scipy import special

def voigt(x,y):
     """voigt profile"""
     #return special.wofz((x+1j*y)/sqrt(2)).real/sqrt(2*pi)
     return special.wofz((x+1j*y)).real/sqrt(pi)

dopplerWidth = 0.0106    
fsr = 0.0205747
B0 = 0.7134
X0 = 7822.98
Y0 = 1.2
A0 = 0.589/voigt(0.0,Y0)
xfine = X0 + fsr*linspace(-6.0,6.0,1001)
xi = X0 + fsr*linspace(-5.0,5.0,11)
f = lambda(x): A0*voigt((x-X0)/dopplerWidth,Y0) + B0
figure(1); plot(xfine, f(xfine), xi, f(xi), 'o')
figure(2);
s1, s2, s3 = 0.25, 0.5, 0.75
t1, t2, t3 = 0.25, 0.6, 0.75
plot(xfine, f(xfine), fMeas, f(fMeas), 'o')
plot(fMeas+s1*fsr, f(fMeas+s1*fsr), 'o')
plot(fMeas+s2*fsr, f(fMeas+s2*fsr), 'o')
plot(fMeas+s3*fsr, f(fMeas+s3*fsr), 'o')

fMock = concatenate((fMeas, fMeas+s1*fsr, fMeas+s2*fsr, fMeas+s3*fsr))
uMock = concatenate((f(fMeas),f(fMeas+t1*fsr), f(fMeas+t2*fsr), f(fMeas+t3*fsr)))
```

```python
# LaserUsed has (0-origin) actual laser in lowest two bits and (0-origin) virtual laser in the next 3 bits
# (((vLaserNum - 1) & 7)<<2) + ((aLaserNum - 1) & 3)
```

```python
# Truncate the H5 file to only contain 28*115 ringdowns
from tables import *

with openFile("RD_1358856226743.h5","r") as h5:
    rd = h5.root.rdData.read()
    h5.close()

period = 115
nPeriods = 28
rd = rd[:nPeriods*period]
cd = cd[:nPeriods/4]
cd["RDDataSize"] = 4*period
sd = sd[4::4]
nMock = 4*period

rd["subschemeId"] = 61
for i in range(nPeriods/4):
    s = i * nMock
    rd["uncorrectedAbsorbance"][s:s+nMock] = uMock +0.000*standard_normal(uMock.shape)
    rd["waveNumber"][s:s+nMock] = fMock
    rd["waveNumberSetpoint"][s:s+nMock] = fMock
    rd["laserUsed"][s:s+nMock] = concatenate((period*[0],period*[1<<2],period*[2<<2],period*[3<<2]))
    rd["subschemeId"][s+nMock-1] |= 32768
    
with openFile("out.h5","w") as h5:
    h5.createTable("/","rdData",rd)
    h5.createTable("/","controlData",cd)
    h5.createTable("/","sensorData",sd)
    h5.close()
    
```
